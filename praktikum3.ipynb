```python
def anagram(kata1, kata2):
    if sorted(kata1) == sorted(kata2):
        return True
    else:
        return False
    
print(anagram("kasur", "rusak"))
print(anagram("listen", "silent"))
print(anagram("ayam", "ikan"))
```

```python
def factorial(n):
    return 1 if n == 0 else n * factorial(n-1)
print(factorial(10))
print(factorial(5))
print(factorial(0))
```

Test Case 5

```python
def palindrome(kata):
    cleaned = ''.join(c.lower() for c in kata if c.isalnum)
    return cleaned == cleaned[::-1]

print (palindrome('racecar'))
print (palindrome('python'))
print (palindrome('habibah'))
```

Test Case 6

```python
def is_armstrong(num) :
    digits = [int(d) for d in str(num)]
    power = len(digits)
    return num == sum(d ** power for d in digits)

print(is_armstrong(153))
print(is_armstrong(370))
print(is_armstrong(123))
```

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.grades = []

    def add_grade(self, grade):
        self.grades.append(grade)
        return f"grade {grade} added."

    def get_average(self):
        if len(self.grades) == 0:
            return "no grades available."
        avg = sum(self.grades) / len(self.grades)
        return f"average grade: {avg}"
    

student = Student("Alice")
print(student.add_grade(90))  
print(student.add_grade(80))  
print(student.add_grade(70))  
print(student.get_average())  
```
